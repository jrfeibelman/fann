```python
 print("Formula used in ____.pct_change():")
Math(r'r_t = \frac{p_{t}}{p_{t-1}} - 1')
```

```python
daily_close = aapl[['Adj Close']]

# Daily returns
daily_pct_change = daily_close.pct_change()

# Replace NA values with 0
daily_pct_change.fillna(0, inplace=True)

print("Daily Returns:\n")
print(daily_pct_change)

# Daily log returns : used for better insight into growth of returns over time
# Equivalent to: daily_log_returns_shift = np.log(daily_close / daily_close.shift(1)))
daily_log_returns = np.log(daily_close.pct_change()+1)

print("\nDaily LOG Returns:\n")
print(daily_log_returns)

# Resample `aapl` to business months, take last observation as value 
monthly = aapl.resample('BM').apply(lambda x: x[-1])

# Calculate the monthly percentage change
print("\nMonthly Returns Percent Change:\n")
print(monthly.pct_change())

# Resample `aapl` to quarters, take the mean as value per quarter
quarter = aapl.resample("4M").mean()

# Calculate the quarterly percentage change
print("\nQuarterly Returns Percent Change:\n")
print(quarter.pct_change())

# Plot the distribution of `daily_pct_change`
daily_pct_change.hist(bins=50)
plt.show()

# Pull up summary statistics
print(daily_pct_change.describe())

# Calculate the cumulative daily returns : useful to determine value of investment at regular intervals
print("\Cumulative Daily Returns:\n")
cum_daily_return = (1 + daily_pct_change).cumprod()
print(cum_daily_return)

# Plot the cumulative daily returns
cum_daily_return.plot(figsize=(12,8))
plt.show()

print("\Cumulative Monthly Returns:\n")
# Resample the cumulative daily return to cumulative monthly return 
cum_monthly_return = cum_daily_return.resample("M").mean()
print(cum_monthly_return)
```

## Ordinary Least-Squares Regression (OLS)

```python
print("OLS to minimize squared residuals")
Math(r'$$\hat{y_i} = \hat\alpha + \hat\beta * x_i + \hat\epsilon_i$$ \qquad\qquad \sum_1^n(Y_i-\hat{Y_i})^2=\sum_1^n(\hat\epsilon_i)^2')
```

```python
stock1 = "AMZN"
stock2 = "NFLX"

# Isolate the adjusted closing price
all_adj_close = data[['Adj Close']]

# Calculate the returns 
all_returns = np.log(all_adj_close / all_adj_close.shift(1))

# Isolate the AAPL returns 
aapl_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == stock1]
aapl_returns.index = aapl_returns.index.droplevel('Ticker')

# Isolate the GOOG returns
goog_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == stock2]
goog_returns.index = goog_returns.index.droplevel('Ticker')

# Build up a new DataFrame with AAPL and GOOG returns
return_data = pd.concat([aapl_returns, goog_returns], axis=1)[1:]
return_data.columns = [stock1, stock2]

# Add a constant 
X = sm.add_constant(return_data[stock1])

# Construct & Print the model
model = sm.OLS(return_data[stock2],X).fit()
print(model.summary())

# How to read the OLS Results:
# Dep. Variable: which variable is the response in model
# Model: what model you are using to fit (OLS)
# Method: How parameters of model were calculated
# R-squared: coefficient of determination, score indicates how well regression approximates real data points with a percentage. 0 means model explains none of variability of response data around its mean.
# Adj R-squared: R-squared adjusted for DF residuals
# F-statistic: how significant fit is, calculated by dividing mean squared error of model by mean squared error of residuals
# Prob (F-statistic): indicates probability that you get the result of F-statistic given the null hypothesis that they are unrelated
# Log-likelihood: indicates log of likelihood function
# AIC: Akaike Information Criterion: adjusted log-likelihood based on number of observations & model complexity
# BIC: Bayesian Information Criterion: like AIC, but penalizes models with more parameters more severly

# coef: estimated value of coefficient
# std err: standard error of coefficient estimate
# t: t-statistic value used to measure how statistically significant a coefficient is
# P > |t|: indicates the null-hypothesis that the coefficient=0 is true. If less than the confidence level, often 0.05, indicates a statistically significant relationship between term and the response.
# Omnibus: Omnibus D'Angostino's test provides combined statistical test for presence of skewness and kurtosis
# Prob(Omnibus): Omnibus metric turned into a probability
# Skew: measures symmetry of data about mean
# Kurtosis: gives indication of shape of distribution as compares the amount of data close to mean with those far away from mean (in tails)
# Durbin-Watson: test for presence of autocorrelation
# Jarque-Bera: test of skewness and kurtosis
# Prob(JB): Jarque-Bera as a probability
# Cond. No.: tests multicollinearity

# Plot returns of AAPL and MSFT
print("\nPlot of Ordinary Least-Squares Regression:")
plt.plot(return_data[stock1], return_data[stock2], 'r.')

# Add an axis to the plot
ax = plt.axis()

# Initialize `x`
x = np.linspace(ax[0], ax[1] + 0.01)

# Plot the regression line
plt.plot(x, model.params[0] + model.params[1] * x, 'b', lw=2)

# Customize the plot
plt.grid(True)
plt.axis('tight')
plt.xlabel(stock1 + ' Returns')
plt.ylabel(stock2 + ' Returns')
plt.show()

# Rolling Correlation
print("Plot of Rolling Correlation:")
# Plot the rolling correlation
return_data[stock2].rolling(window=252).corr(return_data[stock1]).plot()
plt.show()
```

# Trading Strategy

```python
# Fixed plan to go long or go short --> momentum vs. reversion strategy

# Momentum Strategy (divergence/trend trading): believe movement of quantity will continue in current direction and that stock has momentum upwards or downwards to detect and exploit
# Moving Average Crossover: prie of an asset moves from one side of a moving avg to the other. This crossover represents a change in momentum & is used as a point of making the decision to enter/exit market
# Dual Moving Average Crossover: short-term avg crosses a long-term avg, used to identify that momentum is shifting in the direction of the short-term avg. Buy when short-term avg crosses long-term avg and rises above it. Sell when short-term avg crosses long-term avg and falls below it.
# Turtle Trading: buy futures on 20-day high and sell on 20-day low

# Reversion Strategy (convergence/cycle trading): believe movement of a quantity will reverse
# Mean Reversion Strategy: believe a stock will return to its mean and can exploit it when deviates from it
# Pairs Trading Mean-Reversion: if two stocks can be identified that have a relatively high correlation, the change in difference in price between the two stocks can be used to signal trading events if one of the two moves out of correlation with the other. If correlation decreases, stock with a higher price considered to be in a short position and should be sold because the higher priced stock will return to the mean. The lower-priced stock will be in a long position because the price will rise as the correlation will return to normal.
# Forcasting Strategy: predicts direction/value of a stock in subsequent future time periods based on certain historical factors
# High-Frequency Trading: Exploits sub-millisecond market microstructure
```

#### 1) Implementation (Moving Average Crossover)

```python
# Moving Average Crossover: create two separate simple moving averages of a time series with different lookback periods (ie. 40 & 100 days)
# If short moving average exceeds long moving average: enter/go long (buy)
# If long moving average exceeds short moving average: exit (sell)

target_name = 'NFLX'
target = data.loc[target_name]

# 1) define two different lookback periods (short & long windows)
short_window = 40
long_window = 200

# 2) make empty signals DataFrame (copy index of stock data to start calculating daily buy or sell signals for data)
signals = pd.DataFrame(index=target.index)
signals['signal'] = 0.0

# 3) create long & short simple moving average over the short window (min_periods is min # observations in window required to have a value, center=False so labels are not set at center of window)
signals['short_mavg'] = target['Adj Close'].rolling(window=short_window, min_periods=1, center=False).mean()
signals['long_mavg'] = target['Adj Close'].rolling(window=long_window, min_periods=1, center=False).mean()

# 4) create signals for when short moving avg crosses long moving avg, but only for the period greater than the shortest moving avg window
# add [short_window:] to comply with condition 'only for period greater than the shortest moving avg window'. If true, value in signal is overwritten from 0 to 1
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1.0, 0.0)   

# 5) take difference of signals to enerate trading orders (to distinguish between long and short positions)
signals['positions'] = signals['signal'].diff()

# print(signals)

# Create Plot of short & long moving avg
fig = plt.figure()

# Add a subplot and label for y-axis
ax1 = fig.add_subplot(111,  ylabel='Price in $')

# Plot closing price
target['Adj Close'].plot(ax=ax1, color='r', lw=2.)

# Plot the short and long moving averages
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

# Plot the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, 
         signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='m')
         
# Plot the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, 
         signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='k')
         
# Show the plot
plt.show()

# print all buy and sell orders at their price
# pre_backtest = pd.concat([signals, target['Adj Close']], axis=1)[1:]
# backtest = pre_backtest.where(pre_backtest['positions'] != 0)
# print(backtest.to_string())
```

#### 2) Backtesting Trading Strategy

```python
# 4 Essential Components (can include more optional ones):
# Data Handler: interface to a set of data
# Strategy: generates a signal to go long or short based on data (implemented above)
# Portfolio: generates orders and manages Profit & Loss (PnL)
# Execution Handler: sends order to broker and receives 'fills' or signals that the stock was bought or sold

# Python Tools:
# Pandas, NumPy, SciPy for vectorization, optimization, linear algebra
# Scikit-Learn: machine learning library to forecast strategies for regression and classification
# statsmodel: use statistical library for (for example) time series analysis like Ordinary Least-Squares Regression
# IbPy: API for interactive Brokers online trading system - request stock ticker data & submit stock orders
# ZipLine: all-in-one backtesting framework that powers Quantopian
```

```python
# Set the initial capital & Create a DataFrame `positions`
initial_capital= float(10000.0)
positions = pd.DataFrame(index=signals.index).fillna(0.0)

# Buy a 100 shares of Apple on days that the short moving avg crosses long moving avg (for period greater than shortest moving avg window)
target_shares = 100
positions[target_name] = target_shares*signals['signal']   
  
# Initialize the portfolio with value owned   
portfolio = positions.multiply(target['Adj Close'], axis=0)

# Store the difference in shares owned 
pos_diff = positions.diff()

# Backtesting:
# Add `holdings` to portfolio
portfolio['holdings'] = (positions.multiply(target['Adj Close'], axis=0)).sum(axis=1)

# Add `cash` to portfolio
portfolio['cash'] = initial_capital - (pos_diff.multiply(target['Adj Close'], axis=0)).sum(axis=1).cumsum()   

# Add `total` to portfolio
portfolio['total'] = portfolio['cash'] + portfolio['holdings']

# Add `returns` to portfolio
portfolio['returns'] = portfolio['total'].pct_change()

# Print the lines of `portfolio`
# print(portfolio.to_string())

# Create a Plot
fig = plt.figure()
ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')

# Plot the equity curve in dollars
portfolio['total'].plot(ax=ax1, lw=2.)

ax1.plot(portfolio.loc[signals.positions == 1.0].index, portfolio.total[signals.positions == 1.0], '^', markersize=10, color='m')
ax1.plot(portfolio.loc[signals.positions == -1.0].index, portfolio.total[signals.positions == -1.0],'v', markersize=10, color='k')

plt.show()

print("Gain/Loss using $", initial_capital, "for", target_shares,"share increments: ", portfolio['total'][-1] - initial_capital)

```

#### 3) Evaluate Trading Strategy

```python
print("Sharpe Ratio tells us whether portfolio returns are because of smart investments or lots of risks")
print("Sharpe Ratio = Portfolio Excess Returns / Portfolio Volatility")

Math(r' = \frac{E[{R_P}] - r_f}{\sigma_P}')
```

```python
# Note: This does NOT include the risk free rate!
# Use this ratio by comparing it to that of other stocks (typically by extending trading strategy to more data of other companies)

# Isolate the returns of your strategy
returns = portfolio['returns']

# annualized Sharpe ratio for a 252 day window
sharpe_ratio = np.sqrt(252) * (returns.mean() / returns.std())

# Print the Sharpe ratio
print(sharpe_ratio)
```

```python
# Max Drawdown score indicates the risk of a portfolio chosen based on a certain strategy
 
# Define a trailing 252 trading day window
window = 252

# Calculate the max drawdown in the past window days for each day 
rolling_max = target['Adj Close'].rolling(window, min_periods=1).max()
daily_drawdown = target['Adj Close']/rolling_max - 1.0

# Calculate the minimum (negative) daily drawdown
# Note: set min_periods to 1 because want to let the first 252 days data have an expanding window
max_daily_drawdown = daily_drawdown.rolling(window, min_periods=1).min()

# Plot the results
daily_drawdown.plot()
max_daily_drawdown.plot()

# Show the plot
plt.show()
```
